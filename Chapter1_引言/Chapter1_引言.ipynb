### 1 引言
#### 1.1 日常中的机器学习
#### 1.2 机器学习中的关键组件
##### 1.2.1 数据
- 数据集（Dataset）
- 样本（Sample）
- 特征向量（Eigenvalue）
##### 1.2.2 模型
- 深度学习（Deep Learning）：模型由神经网路错综复杂交织，包含层层数据转换。
##### 1.2.3 目标函数
- 目标函数（Objective Function）：定义模型的优劣程度，越小越好。
- - 损失函数（Loss Function）
- - 平方误差（Squared Error）：最常见的损失函数
- 过拟合（Overfitting）
##### 1.2.4 优化算法
- 梯度下降（Gradient Descent）

#### 1.3 各种机器学习问题
##### 1.3.1 监督学习
- 监督学习（Supervised Learning）：在给定输入特征的情况下预测标签。
- 样本（Sample）：特征-标签对。
1. 回归（Regression）：解决“有多少”的问题。
2. 分类（Classification）：解决“是不是”的问题。
3. 标注问题
- 多标签分类（Multi-Label Classification）：学习预测不互相排斥的类别。
4. 搜索
5. 推荐系统（Recommender System）
6. 序列学习
##### 1.3.2 无监督学习
- 无监督学习（Unsupervised Learning）：不含目标的机器学习。
- - 聚类（Clustering）：在没有标签的情况下给数据分类。
- - 主成分分析（Principal Component Analysis）：找到少量参数准确捕捉数据的线性相关性
- - 因果关系（Causality）和概率图模型（Probabilistic Graphical Model）：描述观察到的数据根本原因
- - **生成对抗网络（Generative Adversarial Networks, GCN）**
##### 1.3.3 与环境互动
##### 1.3.4 强化学习
- 强化学习（Reinforcement Learning）：可将监督学习转化为强化学习
- - 智能体（Agent）：在一系列的时间步骤上与环境交互，在每个特定时间点，Agent从环境中接收观察（Observation），选择一个动作（Action），然后通过某种机制（执行器）将其传输回环境，最后Agent从环境中获得奖励（Reward），开始新一轮循环。
